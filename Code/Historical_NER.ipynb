```python
def sent_full_text(text):
  doc= nlp(text)
  sentences = []
  for sent in doc.sents:
    s=sent.text
    sentences.append(s)
  return sentences
```

```python
#chat
def split_into_sentences(text, max_chars=500):
    doc = nlp(text)
    sentences = []
    for sent in doc.sents:
        s = sent.text.strip()
        if len(s) > max_chars:
            # hard fallback split
            chunks = [s[i:i+max_chars] for i in range(0, len(s), max_chars)]
            sentences.extend(chunks)
        else:
            sentences.append(s)
    return sentences
```
